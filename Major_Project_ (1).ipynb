```python
#VIEWING THE DATASET:

import pandas as pd
df=pd.read_csv('Air_Quality.csv')
df.head(5)



```

```python
#COLUMN INFORMATION

df.info()
```

```python
# CALCULATING MEAN AND ROUND OFF TO 2 DECIMAL PLACES
df.isna().mean().round(4)*100
```

```python
#This code is for making a bar chart that shows the average pollution levels in different cities.
#It selects the top 10 cities with the Least pollution levels.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame containing air quality data
# Replace 'city' and 'pollutant_avg' with your actual column names
# if they are different.

# Group by city and calculate average PM2.5 levels
city_pm25_avg = df.groupby('city')['pollutant_avg'].mean().sort_values()

# Get the top 10 least polluted cities
bottom_10_cities = city_pm25_avg.head(10)

# Create a bar plot using Seaborn
plt.figure(figsize=(10, 6))
palette = sns.color_palette("viridis", n_colors=len(bottom_10_cities))
bottom_10_cities.plot(kind='bar', color=palette)
plt.title('Top 10 Least Polluted Cities (PM2.5 Levels)')
plt.xlabel('City')
plt.ylabel('Average PM2.5 Level')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
#This code is for making a bar chart that shows the average pollution levels in different cities.
#It selects the top 10 cities with the highest pollution levels.
# Assuming 'df' is your DataFrame containing air quality data
# Replace 'city' and 'pollutant_avg' with your actual column names
# if they are different.

# Group by city and calculate average PM2.5 levels
city_pm25_avg = df.groupby('city')['pollutant_avg'].mean().sort_values(ascending=False)

# Get the top 10 most polluted cities
top_10_cities = city_pm25_avg.head(10)  # Assigning the top 10 to top_10_cities

plt.figure(figsize=(10, 6))
palette = sns.color_palette("dark:salmon_r", n_colors=len(top_10_cities))
top_10_cities.plot(kind='bar', color=palette)
plt.title('Top 10 Most Polluted Cities (PM2.5 Levels)')
plt.xlabel('City')
plt.ylabel('Average PM2.5 Level')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# We have some information about different states and their average pollution levels.
# We want to find the top 5 states with the highest pollution.
# Filter data for PM2.5 pollutant (you can change to other pollutants)
pm25_data = df[df['pollutant_id'] == 'PM2.5']

# Group by state and calculate the average PM2.5 levels
state_pm25_avg = pm25_data.groupby('state')['pollutant_avg'].mean().sort_values(ascending=False)

# Get the top 5 most polluted states
top_5_states = state_pm25_avg.head(5)

# Create a bar plot using Seaborn
plt.figure(figsize=(10, 6))
palette = sns.color_palette("Reds", n_colors=len(top_5_states))
top_5_states.plot(kind='barh', color=palette)
plt.title('Top 5 Most Polluted States (PM2.5 Levels)')
plt.xlabel('State')
plt.ylabel('Average PM2.5 Level')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
#We have some information about different states and their average pollution levels.
#We want to find the top 5 states with the Least pollution.
# Filter data for PM2.5 pollutant
pm25_data = df[df['pollutant_id'] == 'PM2.5']

# Group by state and calculate the average PM2.5 levels
state_pm25_avg = pm25_data.groupby('state')['pollutant_avg'].mean().sort_values()

# Get the top 5 states with the lowest average PM2.5 levels
least_polluted_states = state_pm25_avg.head(5)

# Plot the results using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=least_polluted_states.index, y=least_polluted_states.values, palette="Greens")
plt.title('Top 5 Least Polluted States (PM2.5 Levels)')
plt.xlabel('State')
plt.ylabel('Average PM2.5 Level')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# We made a table that shows how much pollution each state has,

state_pollution_table= pd.crosstab(index=df['state'], columns=df['pollutant_id'])
```

```python
# and then we made a bar graph to show it more clearly
state_pollution_table.plot(kind='bar', figsize=(10, 10), stacked=True)
```

```python
# We are using a tool called Plotly Express to create a map.
# We are showing the pollution levels in different states of India.

import plotly.express as px
fig = px.choropleth(
    df,
    geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
    featureidkey='properties.ST_NM',
    locations='state',
    color='pollutant_avg',
    color_continuous_scale='temps',
)
fig.update_geos(fitbounds="locations", visible=False)
fig.show()
```
